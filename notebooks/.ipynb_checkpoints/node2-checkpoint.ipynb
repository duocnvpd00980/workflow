```python
import asyncio
import pprint
import json

from agent_os.system.bus.main_bus import MainBus
from agent_os.nodes_library.node_supervisor.adapter_supervisor import node_supervisor


def safe_serialize(obj):
    if hasattr(obj, "model_dump"):
        return obj.model_dump()
    if hasattr(obj, "dict"):
        return obj.dict()
    if isinstance(obj, dict):
        return obj
    return str(obj)


async def run_test():

    raw_mock_state = {
        "user_input": "xin chào",
        "input_guard": {
            "execution_id": "7b378ecf-5e53-41ee-9791-fcd313ed56d8",
            "trace_id": "a79df22f-6c72-4044-9a01-f4c1c3e1c3fa",
            "node_id": "input_guard",
            "payload": {
                "status": "SUCCESS",
                "text": "xin chào",
                "state": {
                    "is_safe": True,
                    "sanitized_text": "xin chào",
                    "risk_category": "NONE",
                    "user_input": "xin chào",
                    "history": [
                        {"role": "user", "content": "xin chào"},
                        {"role": "assistant", "content": "acknowledged"}
                    ],
                    "user_profile": {
                        "user_id": "mock_user_001",
                        "persona": "marketing_user"
                    },
                    "relevant_episodes": [
                        {
                            "event": "previous_query",
                            "summary": "user asked about marketing"
                        }
                    ],
                    "evaluation_feedback": ""
                }
            },
            "timestamp": 1779374134.175927
        },
        "supervisor": {
            "execution_id": "c7020f7c-748c-4ed4-b742-a90a88ae51da",
            "trace_id": "ba9499a9-7814-424b-aad7-b2bd9a8c1aca",
            "node_id": "supervisor",
            "payload": {
                "status": "FAILED",
                "state": {
                    "next_action": "end",
                    "process_completed": False
                },
                "metrics": {
                    "confidence_score": 0.0
                },
                "context": {
                    "topology_error": "[node_supervisor] Missing INPUT_GUARD upstream frame"
                },
                "error": "[node_supervisor] Missing INPUT_GUARD upstream frame"
            },
            "timestamp": 1779374134.17788
        },
        "errors": []
    }

    # =========================================================
    # ÉP KIỂU MAINBUS
    # =========================================================
    mock_state = MainBus.model_validate(raw_mock_state)

    print("🚀 Running node_supervisor...")

    # =========================================================
    # EXECUTE NODE
    # =========================================================
    bus_frame = await node_supervisor(state=mock_state)

    # =========================================================
    # OUTPUT
    # =========================================================
    print("\n================ RESULT (JSON) =================")

    print(
        json.dumps(
            safe_serialize(bus_frame),
            indent=2,
            ensure_ascii=False
        )
    )


# =========================================================
# RUN (JUPYTER SAFE)
# =========================================================
await run_test()
```
