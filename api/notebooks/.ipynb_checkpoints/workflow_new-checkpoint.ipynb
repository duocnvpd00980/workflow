```python
# ==============================================================================
# CELL 1: DJANGO BOOTSTRAP — chạy cell này TRƯỚC TIÊN, mọi cell khác phụ thuộc
# ==============================================================================
import sys
import os

os.chdir("/home/duoc/workflow/api")
sys.path.insert(0, "/home/duoc/workflow/api") 


from langgraph.graph import StateGraph, START, END
from langgraph.checkpoint.memory import MemorySaver
from langgraph.types import RetryPolicy
from app.core.main_bus import MainBus
from app.core.registry import BusRegistry
from app.core.shields.shield_faults import PipelineError
from app.core.shields.shield_node import Node


# =========================================================
# MAPPING: CORE AGENT FLOW (Theo sơ đồ Mermaid)
# =========================================================

from app.nodes_library.node_final_response.adapter_final_response import node_final_response
from app.nodes_library.node_human_review.adapter_human_review import node_human_review
from app.nodes_library.node_input_guard.adapter_input_guard import node_input_guard
from app.nodes_library.node_output_guard.adapter_output_guard import node_output_guard
from app.nodes_library.node_cache_layer.adapter_cache_read import node_cache_read
from app.nodes_library.node_cache_layer.adapter_cache_write import node_cache_write
from app.nodes_library.node_heuristic_router.adapter_heuristic_router import node_heuristic_router
from app.nodes_library.node_knowledgebase.adapter_knowledgebase import node_knowledgebase
from app.nodes_library.node_relevance_check.adapter_relevance_check import node_relevance_check
from app.nodes_library.node_fallback_search.adapter_fallback_search import node_fallback_search
from app.nodes_library.node_generation.adapter_generation import node_generation

```

```python



# =========================================================
# BOARD
# =========================================================

optimized_board = StateGraph(MainBus)

# =========================================================
# CORE AGENT FLOW
# =========================================================


# =============================================================================
# 2. ĐĂNG KÝ CÁC NODE (Khớp chuẩn xác tên hàm và định danh hệ thống của cậu)
# =============================================================================
Node(BusRegistry.IG,  node_input_guard).mount(optimized_board)       # Node 1: Kiểm tra an toàn
Node(BusRegistry.RO,  node_heuristic_router).mount(optimized_board)   # Node 2: Bộ định tuyến từ khóa
Node(BusRegistry.CR,  node_cache_read).mount(optimized_board)       # Node 3: Bộ nhớ đệm L1/L2
Node(BusRegistry.CW,  node_cache_write).mount(optimized_board)       # Node 3: Bộ nhớ đệm L1/L2
Node(BusRegistry.FR,  node_final_response).mount(optimized_board)    # Node 8: Trả kết quả UI
Node(BusRegistry.FB,  node_fallback_search).mount(optimized_board)   # Node THÊM MỚI: Khối cứu cánh Fallback

Node(BusRegistry.KLB, node_knowledgebase).mount(optimized_board)     # Node 4: Quét FAISS lấy Context
Node(BusRegistry.RC,  node_relevance_check).mount(optimized_board)   # Node 5: Check rác tài liệu
Node(BusRegistry.GEN, node_generation).mount(optimized_board)    # Node 6: Gọi LLM sinh câu trả lời
Node(BusRegistry.OG,  node_output_guard).mount(optimized_board)       # Node 7: Kiểm tra đầu ra

# =============================================================================
# 3. THIẾT LẬP ĐƯỜNG ĐI TUYẾN TÍNH - KHỚP 100% MÃ MERMAID CỦA CẬU
# =============================================================================

optimized_board.add_edge(START, BusRegistry.IG)
optimized_board.add_edge(BusRegistry.IG, BusRegistry.RO)
optimized_board.add_edge(BusRegistry.RO, BusRegistry.CR)
optimized_board.add_conditional_edges(
    BusRegistry.CR,
    lambda state: state.route("cache_read", default="miss"),
    {
        "hit": BusRegistry.FR,
        "miss": BusRegistry.KLB,
        "end": END,
    },
)
optimized_board.add_edge(BusRegistry.KLB, BusRegistry.RC)
optimized_board.add_conditional_edges(
    BusRegistry.RC,
    lambda state: state.route("relevance_check", default="low_rel"),
    {
        "high_rel": BusRegistry.GEN,
        "low_rel": BusRegistry.FB,
        "end": END,
    },
)
optimized_board.add_edge(BusRegistry.GEN, BusRegistry.OG)
optimized_board.add_edge(BusRegistry.OG, BusRegistry.CW)
optimized_board.add_conditional_edges(
    BusRegistry.OG,
    lambda state: state.route("output_guard"),
    {
        "cache": END,       
        "skip_cache": END,  
    },
)

optimized_board.add_edge(BusRegistry.OG, BusRegistry.FR)
optimized_board.add_conditional_edges(
    BusRegistry.FB,
    lambda state: state.route("fallback_search"),
    {
        "tool_executed": BusRegistry.RC,
        "error": BusRegistry.FR,
    },
)
optimized_board.add_edge(BusRegistry.FR, END)

# =========================================================
# COMPILE
# =========================================================


main_v7 = optimized_board.compile(
    checkpointer=MemorySaver()
)

# =========================================================
# DEBUG
# =========================================================

if __name__ == "__main__":
    print("\n===== MAINBOARD STRUCTURE =====")

    graph = main_v7.get_graph()

    for node in graph.nodes:
        print(f"Node: {node}")

    print("\n✅ MAINBOARD READY")
```

```python
import os, time, psutil
from IPython.display import display, Markdown, JSON, HTML
from app.graph import main_v7

async def invoke(graph, message: str, thread_id: str = "default"):
    proc   = psutil.Process(os.getpid())
    before = proc.memory_info().rss / 1024 / 1024
    t0     = time.perf_counter()

    try:
        result = await graph.ainvoke(
            {"user_input": message, "language": "vi", "budget_limit": 2.0},
            config={"configurable": {"thread_id": thread_id}},
        )
        elapsed   = round(time.perf_counter() - t0, 2)
        after     = proc.memory_info().rss / 1024 / 1024
        delta_ram = round(after - before, 1)

        display(HTML(f"""
        <style>
            .card {{ display:inline-block;width:140px;margin:6px;padding:14px;
                     border-radius:10px;background:#1e1e2e;color:#cdd6f4;
                     text-align:center;font-family:monospace; }}
            .label {{ font-size:11px;color:#888;margin-bottom:4px; }}
            .value {{ font-size:20px;font-weight:bold;color:#89b4fa; }}
            .unit  {{ font-size:11px;color:#6c7086; }}
        </style>
        <h4 style="font-family:monospace;color:#cdd6f4">📊 "{message}"</h4>
        <div>
            <div class="card"><div class="label">Thời gian</div>
                <div class="value">{elapsed}</div><div class="unit">giây</div></div>
            <div class="card"><div class="label">Delta RAM</div>
                <div class="value">{"+" if delta_ram>=0 else ""}{delta_ram}</div><div class="unit">MB</div></div>
            <div class="card"><div class="label">RAM sau</div>
                <div class="value">{round(after,1)}</div><div class="unit">MB</div></div>
        </div>
        """))

        return {"ok": True, "result": result}

    except Exception as e:
        return {"ok": False, "reason": str(e)}


# ── Chạy ──
result = await invoke(main_v7, "quy định nghỉ phép năm 2025", thread_id="bench-002")

if result["ok"]:
    display(Markdown("### ✅ Kết quả"))
    display(JSON(result["result"]))
else:
    display(Markdown("### ❌ Lỗi"))
    display(Markdown(f"**Lý do:** `{result.get('reason')}`"))

print(f"Status: {'Success' if result['ok'] else 'Failed'}")
```

```python
# WORKFLOW IMAGE

import base64
from IPython.display import Image, display

def draw_cloud_graph(graph):
    # 1. Lấy mã mermaid
    mermaid_code = graph.draw_mermaid()
    
    # 2. Encode sang base64 để gửi qua URL
    graphbytes = mermaid_code.encode("ascii")
    base64_bytes = base64.b64encode(graphbytes)
    base64_string = base64_bytes.decode("ascii")
    
    # 3. Sử dụng service mermaid.ink để lấy ảnh PNG
    display(Image(url=f"https://mermaid.ink/img/{base64_string}"))

# Sử dụng:
draw_cloud_graph(main_v7.get_graph())
```

```python
# =====================================================
# CELL: RUN WORKFLOW (FIXED FULL VERSION)
# =====================================================

# ==============================================================================
# CELL 1: DJANGO BOOTSTRAP — chạy cell này TRƯỚC TIÊN, mọi cell khác phụ thuộc
# ==============================================================================
import sys
import os

sys.path.insert(0, "/app")

# Khai báo settings module trước khi import bất cứ thứ gì của Django
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "config.settings.local")

import django
django.setup()  # Bắt buộc — khởi tạo App Registry trước khi dùng ORM

print("Django setup OK —", django.get_version())
print("Settings module  :", os.environ["DJANGO_SETTINGS_MODULE"])



import uuid
import json
from IPython.display import display

from agent_os.system.runtime.runtime_engine import build_runtime
from agent_os.system.infra.persistence import factory


async def run_ai_pipeline():

    async with factory.get_checkpointer() as db_checkpointer:
        main_app = board.compile(checkpointer=db_checkpointer)

        session_id = f"PROD_{uuid.uuid4().hex[:8]}"
        services = build_runtime()

        initial_state = {
            "session_id": session_id,
            "user_input": "xin chào",
            "language": "vi",
            "budget_limit": 2.0,
        }

        config = {
            "configurable": {
                "thread_id": session_id,
                "services": services,
            }
        }

        print(f"🚀 KHỞI ĐỘNG LUỒNG: {session_id}")
        print("-" * 60)

        # =====================================================
        # STREAM EXECUTION
        # =====================================================
        print("\n🚀 [STARTING GRAPH EXECUTION...]")

        async for event in main_app.astream(initial_state, config, stream_mode="updates"):

            if not event:
                continue

            for node_name, updated_data in event.items():

                print("\n" + "─" * 60)
                print(f"🟣 NODE: 🔹 [{node_name.upper()}]")
                print("─" * 60)

                if isinstance(updated_data, dict) and updated_data:

                    print("📥 STATE UPDATE:")

                    for k, v in updated_data.items():

                        if isinstance(v, list):
                            summary = f"List({len(v)})"

                        elif isinstance(v, dict):
                            summary = f"Dict(keys={list(v.keys())})"

                        else:
                            summary = str(v)

                            if len(summary) > 80:
                                summary = summary[:77] + "..."

                        print(f"   - {k}: {summary}")

                else:
                    print(f"RAW: {str(updated_data)[:120]}")

        print("\n🏁 EXECUTION DONE\n")

        # =====================================================
        # FINAL STATE (3 WAYS DISPLAY)
        # =====================================================
        final_state = await main_app.aget_state(config)

        print("\n================ JSON =================")
        print(json.dumps(final_state.values, indent=2, ensure_ascii=False))



# RUN
await run_ai_pipeline()
```

```python
graph TD
    %% Entry Point
    User([User Request]) --> InputGuard{Input Guard}
    InputGuard -->|Sanitized| Supervisor{Supervisor Agent}
    
    %% Control Plane
    Supervisor -->|Route Task| WorkerA[Worker Agent A]
    Supervisor -->|Route Task| WorkerB[Worker Agent B]
    
    %% Data Plane & Evaluation
    WorkerA --> Eval{Evaluator}
    WorkerB --> Eval
    
    %% Error Handling & Flow Control
    Eval -->|Fail: Revision| Retry[Local Retry Policy]
    Retry -.->|Re-run Task| Supervisor
    
    Eval -->|Pass| Aggregator[Aggregator / Reducer]
    
    %% Human-in-the-loop & State
    Aggregator --> OutputGuard{Output Guard}
    OutputGuard --> HumanReview{Human Approval}
    
    HumanReview -->|Rejected| Supervisor
    HumanReview -->|Approved| SharedState[(Shared State Storage)]
    
    %% Finalization
    SharedState --> FinalResponse[Final Response]
    FinalResponse --> User
```
