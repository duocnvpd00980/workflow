```python
import asyncio
import random
import traceback


async def _scrape_facebook(fb_url: str, headless: bool, business_id: str):
    """
    GIỮ NGUYÊN cấu trúc try/except/finally của bản gốc đã test ok (1000 lần không lỗi).
    Browser tự mở MỚI TINH và tự đóng riêng trong hàm này — không dùng UC_BROWSER global,
    vì node_serp đã đóng hẳn browser của nó trước khi node này chạy.
    """
    browser = None

    try:
        print("🚀 Mở trình duyệt...")
        browser = await uc.start(headless=headless, icon_mode=1)

        # Chuẩn hóa URL để tìm kiếm trực tiếp trên Google
        target_url = fb_url.strip().rstrip("/").lower()
        clean_search_url = target_url.replace("https://", "").replace("http://", "").replace("www.", "")

        print(f"🔍 Vào Google tìm kiếm trực tiếp từ khóa link: {clean_search_url}...")
        
        # Tìm kiếm trực tiếp bằng link sạch theo ý bạn
        page = await browser.get(
            f"https://www.google.com/search?q={clean_search_url}"
        )

        print("⏳ Đợi kết quả xuất hiện...")
        target_element = None

        for attempt in range(30):
            try:
                print(f"\n🔄 Attempt {attempt + 1}/30")

                # Sử dụng selector chuẩn a:has(h3) như bạn yêu cầu
                links = await page.select_all("a:has(h3)")

                if not links:
                    print("⚠️ Chưa có kết quả search.")
                    await asyncio.sleep(0.5)
                    continue

                print(f"🔍 Found {len(links)} search results")

                for i, link in enumerate(links):
                    try:
                        # Dùng get_attribute('href') trực tiếp để KHÔNG BAO GIỜ bị lỗi JSON rỗng (NoneType)
                        href_attr = await link.get_attribute("href")
                        if not href_attr:
                            continue
                            
                        href = href_attr.strip().rstrip("/").lower()
                        print(f"   [{i}] {href}")

                        # Kiểm tra xem link tìm được có chứa link Facebook mục tiêu không
                        if clean_search_url in href:
                            print(f"✅ MATCH CHÍNH XÁC [{i}] {href}")
                            target_element = link
                            break

                    except Exception as e:
                        print(f"❌ Result [{i}] error: {e}")

                if target_element:
                    break

            except Exception as e:
                print(f"❌ Attempt {attempt + 1} failed: {e}")

            await asyncio.sleep(0.5)

        if not target_element:
            print(f"❌ Không tìm thấy homepage khớp với: {fb_url}")
            return None, None

        print(f"✅ Đã tìm thấy homepage: {target_url}")

        try:
            await target_element.scroll_into_view()
        except Exception:
            pass

        await asyncio.sleep(random.uniform(0.34, 0.8))

        # Ép mở thẳng trên Tab hiện tại (Xóa target='_blank') để chống bị chặn Login
        print("🪄 Loại bỏ thuộc tính mở tab mới (target=_blank)...")
        try:
            await page.evaluate(
                """
                (linkElement) => {
                    if (linkElement) {
                        linkElement.removeAttribute('target'); 
                        linkElement.setAttribute('rel', 'noopener noreferrer');
                    }
                }
                """,
                target_element
            )
        except Exception as e:
            print(f"⚠️ Không sửa được attribute (vẫn tiếp tục click): {e}")

        print("🖱️ Click Facebook...")
        await target_element.click()

        # Đợi trang chuyển hướng hoàn toàn trên Tab hiện tại
        await asyncio.sleep(random.uniform(3.5, 5.0))

        print("🔎 Kiểm tra trang Facebook...")
        fb_tab = page 
        current_url = str(fb_tab.url).lower()
        print(f"🔗 URL hiện tại sau khi click: {current_url}")

        # Chặn trường hợp bị đá sang trang Login / Checkpoint công cộng
        if "login" in current_url or "checkpoint" in current_url:
            print("❌ Thất bại: Facebook chặn checkpoint / bắt Login rồi!")
            return None, None

        print("🎉 Đã vào được Facebook hợp lệ.")
        print("📜 Scroll + Expand Posts + Expand Comments...")

        for round_idx in range(2):
            print(f"🔄 Round {round_idx + 1}/10")

            # =====================================================
            # SCROLL
            # =====================================================
            try:
                await fb_tab.evaluate(
                    """
                    window.scrollBy(0, 3000);
                    """
                )
            except Exception:
                pass

            await asyncio.sleep(random.uniform(1.5, 2.5))

        print("📸 Lấy HTML...")
        html = await fb_tab.get_content()
        page_path = f"fb_page_{business_id}.html"

        with open(page_path, "w", encoding="utf-8") as f:
            f.write(html)

        print(f"💾 Đã lưu {page_path} ({len(html):,} ký tự)")

        # =====================================================
        # CLICK "SEE MORE" / "VIEW MORE COMMENTS"
        # =====================================================
        try:
            await fb_tab.evaluate(
                """
                (() => {
                    const targets = [
                        "See more",
                        "View more comments",
                        "View previous comments",
                        "View more replies",
                        "Xem thêm",
                        "Xem thêm bình luận",
                        "Xem phản hồi",
                        "Xem thêm phản hồi"
                    ];

                    const all = document.querySelectorAll("*");

                    for (const el of all) {
                        const text = (el.innerText || "").trim();
                        if (targets.includes(text)) {
                            try {
                                el.click();
                            } catch(e) {}
                        }
                    }
                })();
                """
            )
        except Exception:
            pass

        await asyncio.sleep(random.uniform(0.6, 1.2))

        print("✅ Hoàn tất expand.")
        
        # ==========================================
        # DEBUG DIALOG
        # ==========================================
        await asyncio.sleep(2)
        popup_path = None

        try:
            dialog_count = await fb_tab.evaluate(
                """
                document.querySelectorAll('[role="dialog"]').length
                """
            )
            print("FOUND DIALOGS:", dialog_count)

            popup_html = await fb_tab.evaluate(
                """
                (() => {
                    const dialogs = document.querySelectorAll('[role="dialog"]');
                    if (!dialogs.length) return "NO_DIALOG";
                    return dialogs[dialogs.length - 1].outerHTML;
                })()
                """
            )
            print("POPUP SIZE:", len(popup_html))

            if popup_html and popup_html != "NO_DIALOG":
                popup_path = f"fb_popup_{business_id}.html"
                with open(popup_path, "w", encoding="utf-8") as f:
                    f.write(popup_html)
                print(f"💾 Saved {popup_path}")

        except Exception as e:
            print("Popup debug error:", e)

        return page_path, popup_path

    except Exception as e:
        print("❌ Lỗi:", e)
        traceback.print_exc()
        return None, None

    finally:
        print("⏳ Đóng trình duyệt...")
        try:
            if browser:
                await browser.stop()
        except Exception:
            pass
```
